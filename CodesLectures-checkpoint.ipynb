## Stock prices path

We want to simulate stock paths, based on Geometric Brownian Motion
$$
dS(t) = \mu\cdot S(t)\cdot dt + \sigma\cdot S(t)\cdot dW(t)
$$
But we also seen that 
$$
\begin{cases}
X(t) = \log(S(t))\\
\\
X(t+\Delta t) = X(t) + (\mu-\frac{1}{2}\cdot \sigma^2)\Delta t + \sigma[W(t+\Delta t)-W(t)]
\end{cases}
$$
This is possible thanks to Ito's Lemma.

In other words, we'are able to define the logarithm of stock price and we also know from lecture that Wieman process is a Stochastic process normally distributed. This allow us to consider
$$
W(t+\Delta t)-W(t) = \sqrt{t} \cdot Z \quad , \quad Z \sim N(0,1)
$$

In conclusion, the idea is the following:
1. Simulate $X(t)$ for a $T$ interval (e.g. 1 year) with a certain discretization $\Delta t$ (e.g. 1 day), a fixed interest rate $\mu$ and volatility $\sigma$
2. Iterate for $n$ paths
3. Determine S(t) path by $S(t) = exp[X(t)]$

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.pyplot import figure
import seaborn as sns
import plotly.express as px
from IPython.display import display
```

### Single path

```python
np.random.seed(42)

T = 365 #days
delta = 1 #day
dt = delta/T
S_0 = 100 #initial price
X_0 = np.log(S_0)
mu = 0.05
sigma = 0.04
Xlist = [X_0]

for t in range(delta, T+1):
    Z = np.random.normal(0.0, 1.0)
    X = Xlist[-1] + (mu-0.5*sigma**2)*dt + sigma*np.sqrt(dt)*Z
    Xlist.append(X)

Slist = np.exp(Xlist) #from log to STOCK PRICE


#PLOT
fig, axes = plt.subplots(1, 2, figsize=(12, 5))  # 1 riga, 2 colonne

axes[0].plot(np.arange(0, T+1), Xlist)
axes[0].set_title("Log of price")
axes[0].set_xlabel("T (days)")
axes[0].set_ylabel("X")

axes[1].plot(np.arange(0, T+1), Slist)
axes[1].set_title("Stock price")
axes[1].set_xlabel("T  (days)")
axes[1].set_ylabel("S")

plt.tight_layout()  # Migliora la disposizione
plt.show()
```

### Multi paths

We found just one of the possible paths. It could be useful to obtain a huge number of paths (e.g. 10k) and consider for future decisions a certain percentile of the distribution

```python
np.random.seed(42)

T = 365 #days
delta = 1 #day
dt = delta/T
S_0 = 84147.40 #initial price
X_0 = np.log(S_0)
mu = 0.05
sigma = 0.7

n_paths = 1000

Xlist_mp = np.zeros((n_paths, T+1)) #we create a matrix (with 0 for now): each row correspond to a path, columns are the days -> X[0,0] is the X value the day 0
Xlist_mp[:, 0] = X_0 #for all the paths we fill with X_0 value

#GBM
for p in range(n_paths):
    Z = np.random.normal(0.0, 1.0, T)
    increments = (mu - 0.5 * sigma**2) * dt + sigma * np.sqrt(dt) * Z
    Xlist_mp[p, 1:] = X_0 + np.cumsum(increments)

#From X to S
Slist_mp = np.exp(Xlist_mp)

for i in range(n_paths):
    plt.plot(np.arange(0, T+1), Slist_mp[i, :])


plt.xlabel("T (days)")
plt.ylabel("S")
plt.title(f"Stock price - GBM ({n_paths} simulations)")
plt.show()
```

### Simulation with Real Stock data (Ultimate version)

Now let's try to use some real value
- the volatility 
- the return
$$
\begin{cases}
\text{volatility } = \text{standard deviation of returns}\cdot \sqrt{T} = std[\log(\frac{Price_t}{Price_{t-1}})] \cdot \sqrt{365}\\
\\
\text{log returns } =  \log(\frac{Price_t}{Price_{t-1}})\\
\\
\text{standard deviation of returns } = st.dv(\text{log returns})\\
\end{cases}
$$

Notice that for stock market T = 252, while for cryptocurrencies T = 365

#### Stock data scrape

Since yfinance is often down we use Stooq, a website which contains the majority of stock series stored inside csv file easily readable.

```python
#Stock data scraping: input:ticker and time interval (years) -> output = data
def stock_prices(ticker, time_interval):
    url = f"https://stooq.com/q/d/l/?s={ticker}&i=d"
    df = pd.read_csv(url)
    df['Date'] = pd.to_datetime(df['Date'])
    df_filtered = df[df['Date'] >= pd.Timestamp.today() - pd.DateOffset(years=time_interval)]
    return df_filtered.dropna()
```

```python
stock_prices("AAPL.US", 5) #last 5 years data
```

#### Volatility and Drift calculation

```python
def volatility_drift_calc(ticker, time_interval):
    df_filtered = stock_prices(ticker, time_interval)
    df_filtered["Log Returns"] = np.log(df_filtered['Close'] / df_filtered['Close'].shift(1))
    mean_daily_return = df_filtered["Log Returns"].mean()
    daily_volatility = df_filtered["Log Returns"].std()
    
    T = 365 if "BTC" in ticker else 252
    annual_volatility = daily_volatility*np.sqrt(T)
    drift = (mean_daily_return * T) + (0.5 * annual_volatility**2)
    return print(f"{ticker} volatility over {time_interval} year = {round(annual_volatility*100,2)}%\n{ticker} drift over {time_interval} year = {round(drift*100,2)}%")
```

```python
volatility_drift_calc("AAPL.US", 3)
```

#### STOCK PRICE FORECASTING

The following function recive these quantities
- forecast_T : [days] of total forecasting
- delta : discretization; typically the price is visualized day by day, i.e. delta = 1
- n_paths : number of trajectories 
- ticker
- sigma: volatility
- mu: return
- day_hist: number of days after which we want to see the stock price distribution

```python
def GBM_stockPath(forecast_T, delta, n_paths, ticker, sigma, mu, day_hist):
    dt = delta/365
    df = stock_prices(ticker, 1)
    S_0 = df["Close"].iloc[-1]
    X_0 = np.log(S_0)
    
    
    Xlist_mp = np.zeros((n_paths, forecast_T+1)) #we create a matrix: each row correspond to a path, columns are the days -> X[0,0] is the X value the day 0. It contains zero but will be filled
    Xlist_mp[:, 0] = X_0 #for all the paths we fill with X_0 value

    #GBM
    for p in range(n_paths):
        Z = np.random.normal(0.0, 1.0, forecast_T)  # generate all Z at once
        increments = (mu - 0.5 * sigma**2) * dt + sigma * np.sqrt(dt) * Z
        Xlist_mp[p, 1:forecast_T+1] = Xlist_mp[p, 0] + np.cumsum(increments)

    
    plt.figure(figsize=(10, 6))
    
    #From X to S
    Slist_mp = np.exp(Xlist_mp)

    for i in range(n_paths):
        plt.plot(np.arange(0, forecast_T+1), Slist_mp[i, :])
    
    final_prices = Slist_mp[:,day_hist]

    plt.xlabel("T (days)")
    plt.ylabel(f"{ticker}")
    plt.title(f"{ticker} Price, {forecast_T} days forecasting - GBM ({n_paths} simulations)")
    plt.show()
    
    fig = px.histogram(Slist_mp, 
                       x=final_prices, 
                       histnorm='probability density', 
                       width=900,  
                       height=500, 
                       title=f"{ticker} Prices after {day_hist} days - GBM ({n_paths} simulations)",
                       marginal="box"
                      )
    fig.update_layout(
        title=dict(font=dict(size=16, color="black", family="Arial"), x=0.5), 
        xaxis_title=f"{ticker} - Price",
        yaxis=dict(
            title_font=dict(size=14, color="darkblue"),  
            tickfont=dict(size=12, color="darkblue"),   
            gridcolor="lightgrey"
        ),
        plot_bgcolor="white",
        margin=dict(l=40, r=40, t=60, b=40) 
    )
    fig.show()
```

```python
np.random.seed(42)
GBM_stockPath(365, 1, 1000, "AAPL.US", 0.2, 0.05, 365)
```
